```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
print(df.isnull().sum())
```

```python
import missingno as mis

mis.bar(df)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Carte des Valeurs Manquantes")
plt.show()
```

```python
dff = df.dropna(axis=0)
```

```python
dff.shape
```

```python
dff.dropna(subset=['Embarked'] , axis=0)
print(dff.shape)
```

```python
dff = dff.drop('Cabin' , axis=1)
print(dff.shape)
```

```python
sns.distplot(dff['Age'])
```

```python
df['Age'].fillna(dff['Age'].median() , inplace=True)
```

```python
sns.distplot(df['Age'])
```

```python
df['Embarked'].fillna(dff['Embarked'].mode() , inplace=True)
```
